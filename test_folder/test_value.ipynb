```python
# 제작한 모듈들 임포트
from price.price_processing import *

from backtest.metric import Metric

from strategy.factors.value import *
from strategy.optimize.cross_sectional import *
from strategy.optimize.time_series import *
```

```python
with open('../../part07.-퀀트-트레이딩-실무-실습-파일/chapter3/피클파일/commodity_price.pkl', 'rb') as f:
    commodity_price = pickle.load(f)
    commodity_price.index = pd.to_datetime(commodity_price.index)
    commodity_price.index.name = 'date_time'
```

```python
commodity_price
```

```python
target_rebal = rebal_dates(commodity_price, 'month')
target_rebal
```

```python
rebal_price_df = price_on_rebal(commodity_price, target_rebal)
```

```python
rebal_price_df
```

```python
signal_df = ValueFactor(rebal_price=rebal_price_df, long_only=True).commoditiy_value()
```

```python
signal_df.head()
```

```python
msr_weight = Optimization(signal=signal_df, rebal_price=rebal_price_df, param=12, call_method='msr').run()
gmv_weight = Optimization(signal=signal_df, rebal_price=rebal_price_df, param=12, call_method='gmv').run()
mdp_weight = Optimization(signal=signal_df, rebal_price=rebal_price_df, param=12, call_method='mdp').run()
rp_weight = Optimization(signal=signal_df, rebal_price=rebal_price_df, param=12, call_method='rp').run()
```

```python
cs_weight = Equalizer(signal=signal_df, rebal_price=rebal_price_df, param=12)
beta_weight = cs_weight.beta()
ew_weight = cs_weight.ew()
emv_weight = cs_weight.emv()
```

```python
ew_weight.head()
```

```python
cs_model = {'Beta': beta_weight, 
            'EW': ew_weight, 
            'EMV': emv_weight, 
            'MSR': msr_weight, 
            'GMV': gmv_weight, 
            'MDP': mdp_weight, 
            'RP': rp_weight
            }
```

```python
for name, weight in cs_model.items():
    print(name)
    cum_rets_df = get_cum_rets(get_daily_rets(calculate_portvals(price_df=commodity_price, weight_df=weight)))
    Metric(portfolio=cum_rets_df).print_report()
    print()
```
